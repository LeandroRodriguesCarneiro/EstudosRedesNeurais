```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definindo as funções de ativação e suas derivadas
def step(x):
    return np.where(x >= 0, 1, 0)

def step_derivative(x):
    return np.zeros_like(x)  # Derivada da função degrau é zero em quase todo o domínio

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sig = sigmoid(x)
    return sig * (1 - sig)

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

def softplus(x):
    return np.log(1 + np.exp(x))

def softplus_derivative(x):
    return sigmoid(x)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1, elu(x, alpha) + alpha)
```

```python
# Definindo o intervalo de valores de x
x = np.linspace(-10, 10, 400)

# Calculando os valores das funções de ativação e suas derivadas
y_step = step(x)
y_step_derivative = step_derivative(x)
y_sigmoid = sigmoid(x)
y_sigmoid_derivative = sigmoid_derivative(x)
y_relu = relu(x)
y_relu_derivative = relu_derivative(x)
y_tanh = tanh(x)
y_tanh_derivative = tanh_derivative(x)
y_leaky_relu = leaky_relu(x)
y_leaky_relu_derivative = leaky_relu_derivative(x)
y_softplus = softplus(x)
y_softplus_derivative = softplus_derivative(x)
y_elu = elu(x)
y_elu_derivative = elu_derivative(x)

# Criando os gráficos para as funções de ativação e suas derivadas
plt.figure(figsize=(15, 20))

plt.subplot(7, 2, 1)
plt.plot(x, y_step, label="Step")
plt.title("Step")
plt.grid(True)

plt.subplot(7, 2, 2)
plt.plot(x, y_step_derivative, label="Step Derivative")
plt.title("Step Derivative")
plt.grid(True)

plt.subplot(7, 2, 3)
plt.plot(x, y_sigmoid, label="Sigmoid")
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(7, 2, 4)
plt.plot(x, y_sigmoid_derivative, label="Sigmoid Derivative")
plt.title("Sigmoid Derivative")
plt.grid(True)

plt.subplot(7, 2, 5)
plt.plot(x, y_relu, label="ReLU")
plt.title("ReLU")
plt.grid(True)

plt.subplot(7, 2, 6)
plt.plot(x, y_relu_derivative, label="ReLU Derivative")
plt.title("ReLU Derivative")
plt.grid(True)

plt.subplot(7, 2, 7)
plt.plot(x, y_tanh, label="TanH")
plt.title("TanH")
plt.grid(True)

plt.subplot(7, 2, 8)
plt.plot(x, y_tanh_derivative, label="TanH Derivative")
plt.title("TanH Derivative")
plt.grid(True)

plt.subplot(7, 2, 9)
plt.plot(x, y_leaky_relu, label="Leaky ReLU")
plt.title("Leaky ReLU")
plt.grid(True)

plt.subplot(7, 2, 10)
plt.plot(x, y_leaky_relu_derivative, label="Leaky ReLU Derivative")
plt.title("Leaky ReLU Derivative")
plt.grid(True)

plt.subplot(7, 2, 11)
plt.plot(x, y_softplus, label="Softplus")
plt.title("Softplus")
plt.grid(True)

plt.subplot(7, 2, 12)
plt.plot(x, y_softplus_derivative, label="Softplus Derivative")
plt.title("Softplus Derivative")
plt.grid(True)

plt.subplot(7, 2, 13)
plt.plot(x, y_elu, label="ELU")
plt.title("ELU")
plt.grid(True)

plt.subplot(7, 2, 14)
plt.plot(x, y_elu_derivative, label="ELU Derivative")
plt.title("ELU Derivative")
plt.grid(True)

plt.tight_layout()
plt.show()
```
